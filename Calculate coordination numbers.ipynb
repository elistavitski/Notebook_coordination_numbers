```python
import numpy as np
fn = '/Users/elistavitski/Desktop/global15.xyz'
```

```python
coord = np.loadtxt(fn, skiprows=2,usecols=(1,2,3))
print(len(coord))
```

```python
bonds=[]
for jj in range (len(coord)):
    reduced_coord=coord[:]-coord[jj]
    bond = list(np.sqrt(a**2 + b**2 + c**2) for a, b,c in reduced_coord[:])
    bonds = np.append(bonds, list(round(a,3) for a in bond))
```

```python
b,c = np.unique([bonds], return_counts=True)
print(b)
print(c/len(coord))
```
